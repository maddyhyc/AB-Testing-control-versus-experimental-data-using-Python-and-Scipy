```python
# Import relevant packages

import pandas as pd
import numpy as np
```

```python
# Read file and inspect

audacity = pd.read_csv("audacity.csv")
```

```python
audacity.head(10)
```

```python
#create new column

audacity['action2'] = [1 if x == 'click' else 0 for x in audacity['action']]

audacity.head(10)
```

```python
# Make a 2 x 2 contingency table

table = pd.crosstab(audacity['action2'], audacity['group'], margins=False)
table
```

```python
# package to conduct chi square test for independence

from scipy.stats import chi2_contingency
```

```python
# Using chi2_contingency

chi2_contingency(table)

# chi square table availble at https://web.ma.utexas.edu/users/davis/375/popecol/tables/chisq.html
```
